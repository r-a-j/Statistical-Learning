```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
#Give the values in ascending order
df = pd.DataFrame([[4,3],[3,2],[0,4],[6,4],[6,7],[2,6]])

#Change below for 'single link' and 'complete link'
Z = linkage(df, 'complete') #single #complete
plt.subplots(figsize=(14, 20))

# Plot title
plt.title('Hierarchical Clustering Dendrogram')


# Make the dendrogram
dendrogram(Z, labels=df.values)

# Show the graph
plt.show()
```
